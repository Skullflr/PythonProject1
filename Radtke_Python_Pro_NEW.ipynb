```python
start_time = pd.to_datetime('5:00').time()
end_time = pd.to_datetime('7:00').time()
mask = (df3['time'] > start_time) & (df3['time'] <= end_time)
df = df3.loc[mask]
grouped3 = df.groupby('date')
c = pd.DataFrame(df[df.time.notnull()].groupby('date')['time'].unique())
c_count = list(c.time.value_counts())
print('Number of Early Morning Workouts:',len(c_count))
```
